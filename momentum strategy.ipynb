```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
# getting table elements in wikipedia 
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
tickers = table.Symbol.to_list()
```

```python
len(tickers)
```

```python
df = yf.download(tickers, start='2009-01-01', end='2022-01-01')['Adj Close']
```

```python
df.index = pd.to_datetime(df.index)
```

```python
monthly_ret = df.pct_change().resample('M').agg(lambda x: (x+1).prod() -1)
```

```python
monthly_ret
```

```python
#build past 11 month returns 
# the +1 and -1 is for multiplying percentages
past_11 = (monthly_ret+1).rolling(11).apply(np.prod)-1
```

```python
past_11 
```

```python
formation = dt.datetime(2009,12,31)
```

```python
formation
```

```python
# we cutoff the recent month before get the total result of the positions
end_measurement = formation - MonthEnd(1)
```

```python
end_measurement
```

```python
ret_12 = past_11.loc[end_measurement]
```

```python
ret_12
```

```python
ret_12 = ret_12.reset_index()
```

```python
ret_12
```

```python
ret_12['decile'] = pd.qcut(ret_12.iloc[:,1],10,labels=False, duplicates='drop')
```

```python
ret_12
```

```python
winners = ret_12[ret_12.decile == 9]
losers = ret_12[ret_12.decile == 0]
```

```python
# searching the winners symbols in monthly returns df 
winners_ret = monthly_ret.loc[formation + MonthEnd(1), monthly_ret.columns.isin(winners['index'])]
losers_ret = monthly_ret.loc[formation + MonthEnd(1), monthly_ret.columns.isin(losers['index'])]
```

```python
Momentumprofit = winners_ret.mean() - losers_ret.mean()
Momentumprofit
```

```python
def momentum(formation):
    end_measurement = formation - MonthEnd(1)
    ret_12 = past_11.loc[end_measurement]
    ret_12 = ret_12.reset_index()
    ret_12['decile'] = pd.qcut(ret_12.iloc[:,1], 10 , labels=False, duplicates='drop')
    winners = ret_12[ret_12.decile == 9]
    losers = ret_12[ret_12.decile == 0]
    winners_ret = monthly_ret.loc[formation + MonthEnd(1), monthly_ret.columns.isin(winners['index'])]
    losers_ret = monthly_ret.loc[formation + MonthEnd(1), monthly_ret.columns.isin(losers['index'])]
    Momentumprofit = winners_ret.mean() - losers_ret.mean()
    return Momentumprofit
```

```python
momentum(formation)
```

```python
formation
```

```python
for i in range(12*12):
    print(formation + MonthEnd(i))
```

```python
profits = []
dates = []

for i in range(12*12):
    profits.append(momentum(formation + MonthEnd(i)))
    dates.append(formation + MonthEnd(i))
```

```python
profits
```

```python
SP = yf.download('^GSPC', start=dates[0], end=dates[-1])['Adj Close']
```

```python
SP_monthly = SP.pct_change().resample('M').agg(lambda x: (x+1).prod() -1)
```

```python
SP_monthly
```

```python
final = pd.DataFrame(profits)
```

```python
final['S_P500'] = SP_monthly.values
```

```python
final['excess'] = final.iloc[:,0] - final.iloc[:,1]
```

```python
# list comprehension 
final['outperformed'] = ['Yes' if i > 0 else 'No' for i in final.excess]
```
